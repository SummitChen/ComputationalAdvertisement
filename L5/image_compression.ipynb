```python
import numpy as np
from scipy.linalg import svd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# rbg图像转换为灰度图像
def rgb2gray(rgb):
    return np.dot(rgb, [0.2989, 0.5870, 0.1140])
```

```python
# 提取前百分之portion的特征，对图像进行还原
def get_image_feature(s, portion):
    # 将百分比转换为int型个数值
    k = int(s.shape[0] * portion / 100)
    # 重构s为s_temp,只保留前k个特征值
    s_temp = np.zeros(s.shape[0])
    s_temp[0:k] = s[0:k]
    s = s_temp * np.identity(s.shape[0])
    # 用s_temp，以及p,q重构A
    temp = np.dot(p, s)
    temp = np.dot(temp, q)
    plt.imshow(temp, cmap = plt.cm.gray, interpolation = 'nearest')
    plt.show()
```

```python
# 加载图形并将图像转换为灰度图像
img = mpimg.imread('./data/douluo.jpeg')
A = np.array(img)
plt.imshow(A)
plt.show()
# img = mpimg.imread('./data/scene.jpg')
# A = np.array(image)
# img = color.rgb2gray(io.imread('./data/douluo.jpeg'))
# img = io.imread('./data/douluo.jpeg',as_gray = True)

```

```python
A = rgb2gray(A)
print(A.shape)
# 显示灰度图像
plt.imshow(A, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

```python
p, s, q = svd(A, full_matrices = False)
print(p.shape)
print(s.shape)
print(q.shape)
```

```python
get_image_feature(s, 1)
```
