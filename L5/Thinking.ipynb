### *奇异值分解SVD的原理是怎样的，都有哪些应用场景？*

矩阵分解经常会遇到的一个问题就是所分解的矩阵是非对称的，而不同矩阵分解方法只能解决对称矩阵分解的问题。奇异值分解SVD可以用来解决非对称矩阵的特征值和特征向量分解的问题。简要步骤如下：
- 把要分解的矩阵分别于它自身的转置矩阵进行左乘和右乘等到两个方阵
- 分别这两个方阵进行矩阵分解
- 根据这两个方阵拥有相同的非零特征值这一特性，可将原矩阵分解为左奇异矩阵乘以特征值矩阵在乘以右奇异矩阵。其中左奇异矩阵，特征值以及右奇异矩阵在第二步中得到

SVD可以用于图像压缩，可在只损失少量图像特征的同时大大减少存储图像所需要的存储空间。SVD还可用于完成推荐系统中User和Item的矩阵分解来实现商品推荐。

### *funkSVD, BiasSVD，SVD++算法之间的区别是怎样的？*

传统的SVD分解要求矩阵是稠密的。而推荐系统中的矩阵是稀疏的，且存在大量的缺失值。直接对稀疏矩阵进行分解需要很大的计算量，而且如果对矩阵进行简单的填充的话，又会引入大量的噪音。

FunkSVD的算法思想是

- 不对矩阵直接进行奇异值分解，而是引入k值，将原有的m * n 的矩阵表示成 m * k 的矩阵P与 k * n 矩阵Q相乘的形式；
- 通过解决令损失函数最小的优化问题，来计算矩阵P与Q，优化算法采用的是梯度下降法；
- 损失函数表示为 P 和 Q矩阵相乘所得到的用户评分与实际用户评分之差，再加上正则项。

BiasSVD算法在Funk算法的基础上，引入了偏好项作为评分估计的一部分，编好项是所有记录的整体平均数，用户偏好以及商品偏好三者的相加和;同时在目标函数中也加入用户偏好的平方以及商品偏好的平法作为正则项的一部分。

SVD++算法是在BiasSVD算法的基础上加入了隐式反馈，所谓隐式反馈指的不是用户给的具体评分，而是可能的点击，浏览等行为。

### *矩阵分解算法在推荐系统中有哪些应用场景，存在哪些不足？*

矩阵分解是推荐系统中协同过滤策略的一个重要的方法实现之一，并得到了广泛的应用，其中几个典型的应用场景有
- 个性化页面推荐(“千人千面”)，比如电影电视应用的首页用户推荐
- 目标物品的关联物品推荐场景，可通过矩阵分解获得某一物品的特征向量，然后根据其他物品特征向量到此向量的远近关系来完成推荐
- 目标用户的关联用户推荐场景，将相似用户感兴趣或操作过的物品推荐给目标用户

矩阵分解算法存在的不足主要有：

- 冷启动问题
- 可解释性不强：通过矩阵算法将矩阵分解获得隐式特征，根据隐式特征来估计用户对物品的兴趣水平来进行推荐。但是由于隐式特征没有确定的物理意义，无法对推荐的结果进行解释。只能通过离线或者在线评估来评价算法的效果。

### *item流行度在推荐系统中有怎样的应用？*

1. 在解决冷启动问题方面：

    由于新加入的用户还没有与系统中的物品进行过交互，无法推断出用户的偏好。那么则可以根据物品在整个系统中的流行度进行排序，将流行度（热度）最高的几个物品推荐给用户。
    

2. 流行度高的物品反映了较高的从众行为以及较少的个人偏好：

    根据当前应用目标的不同，比如有些电商网站的目标是打造爆款产品，而婚恋网站更加关注个性化的匹配程度。针对这两种关于流行度不同方向应用，可通过引入权重因子来修正流行度。对于电商网站可以在原有流行度的基础上增加流行商品的流行度，而婚恋网站则倾向于减少过热对象的流行度，来降低群体“跟风效应”的影响。


3. 对某些流行度高的商品进行分析，可以帮助检测恶意刷单行为

    为了提升某售卖商品的在电商系统中的流行度，某些商家可能会采取恶意刷单行为，来提高某商品的流行度，以获得在系统中跟高的推荐可能。可以针对流行商品的用户行为分析，来检测恶意刷单行为。

### *推荐系统的召回阶段都有哪些策略？* 

推荐系统中常用的召回策略有：
1. 以内容为索引的召回，比如某热点文章，商品以及在特定地区，时间发布的新闻或者产品。
2. 以用户为索引的召回，根据某用户的行为来召回，比如该用户最近看过的内容及时长。
3. 以设备为索引的召回，比如根据用户机器设备对应的群体，推荐点击率最高的内容。比如针对iPhone的手机用户，推荐苹果品牌在系统中的明星产品。 
4. 以空间位置为索引的召回，比如中国不同地区的饮食习惯，推荐在相关品类的食品。
