```python
import numpy as np
```

```python
# Normal case
transition_matrix = np.array([[0, 1/2, 1, 0],
                             [1/3, 0, 0, 1/2],
                             [1/3, 0, 0, 1/2],
                             [1/3, 1/2, 0, 0]])
```

![](images/graph.png)

```python
# Leak case where node B is the leaking node
transition_matrix_l = np.array([[0, 0, 1, 0],
                             [1/3, 0, 0, 1/2],
                             [1/3, 0, 0, 1/2],
                             [1/3, 0, 0, 0]])
```

```python
# Sink case where node C is the sinking node
transition_matrix_s = np.array([[0, 1/2, 1, 0],
                                [1/2, 0, 0, 1],
                                [0, 0, 0, 0],
                                [1/2, 1/2, 0, 0]])
```

```python
# initial weight
weight = np.array([1/2, 1/6, 1/6, 1/6])
```

```python
# basic
def work(a, w, s = 100):
    for i in range(s):
        w = np.dot(a, w)
        print("{} - {}".format(i+1, w.round(4)))
    print("*"*100)
```

```python
# with random factor
def random_work(a, w, s=100, d=0.85):
    n = len(w)
    for i in range(s):
        w = (1-d)/n + d*np.dot(a, w)
        print("{} - {}".format(i+1, w.round(4)))
    print("*"*100)
```

```python
# without random factor
work(transition_matrix, weight, 100)
work(transition_matrix_l, weight, 100)
work(transition_matrix_s, weight, 100)
```

```python
# with random factor
random_work(transition_matrix, weight, 100, 0.85)
random_work(transition_matrix_l, weight, 100, 0.85)
random_work(transition_matrix_s, weight, 100, 0.85)
```
