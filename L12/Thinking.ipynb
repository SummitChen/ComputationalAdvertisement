### ***1. 高德地图中的路径规划原理是怎样的？***

交通道路地图可以抽象成带有权重的图结构。图中的节点依照不同的规模层级可以表示一个城市，一个社区，或者一个交通路口；而图中的边则代表这些节点的连接关系。边的权重由边所连接的两个节点的距离表示，也可以更加综合地用乘坐交通工具从一个节点到达另外一个节点的难易程度表示：比如所需花费的时间，以及汽车可能的耗油情况（因堵车而导致的走走停停的现象会增加汽车的耗油）。在假定交通工具（比如一辆普通的汽车）相同的情况下，到达的难易程度会受到路况、道路带宽、道路最高限速情况以及交通流量情况等因素影响。除交通流量以外的其他因素相对固定，属于固定变量；而交通流量则是一个实时变化的指标。预测交通流量的一个可行方法是PageRank, 将节点之间在历史不同时段测得的流量信息抽象成转移矩阵；根据PageRank算法就可以预测出在未来不同时刻通过每个节点的流量情况。根据影响因素的具体信息，就可以计算出难易程度的具体数值。然后采用动态规范方法，就可以在地图中为车辆做路径规划了。

### ***2. football.gml数据集中，美国大学生足球联赛，包括115支球队，被分为12个联盟。为什么使用LPA标签传播进行社区发现，只发现了11个社区？***

联盟的划分和基于LPA标签传播的社区发现采用的分类标准不同，联盟是根据球队的运营从属关系划分的，是一个相对静态的指标。而基于LPA标签传播的方法基于的是球队之间的交互关系（比赛）。那么某些球队与联盟以外的球队比赛的次数可能比同联盟内的球队还多；所以经常在一起比赛的球队就会通过LPA算法，划分到同一个社区内。而由于比赛的基本规则是先在联盟内部进行小组赛，然后是联盟之间的比赛；所以联盟这一特征在LPA社区发现过程中也得到了体现，这就是为什么发现了与联盟数相接近的社区数量，而不是5个社区或者20个社区。

### ***3. 微博采用了类似FaceBook的EdgeRank算法，如果你给微博的信息流做设计，你会如何设计？***

我们将微博的信息流设计问题简化为微博内容的排序问题，即如何将微博用户好友产生的新内容排序和推荐给用户。

这个问题可以采用EdgeRank算法来解决，EdgeRank算法的核心思想是计算每条新事件（发布新文章，状态，图片）对于当前用户的权重E，而权重的计算公式为 E=u*w*d

其中：u表示内容发布者与当前用户的亲密程度，w 表示边权重， d 代表时间衰减因子。

具体来讲， 

- 亲密程度
    - 主要是指用户之间的互动程度，互动越频繁亲密度越高
    - 互动可以是单向的，比如A经常关注B，而B不怎么关注A，则u(A,B)与u(B,A)的值不同。

- 边权重
    - 即针对不同的事件类型，根据不同的规则计算出相应的值；
    - 比如可根据对不同事件的平均点击情况来设置权重值；
    - 还可以根据事件本身的属性，如内容质量，是否为原创内容等。
    - 总来来说，可以综合事件类型和事件属性来计算权重值, 即 w=w_category*w_attribute
 
- 时间衰减因子
    - 即随着时间流逝，事件的重要新在逐渐递减。
    
此外除了，这些因素之外，还可以考虑事件发布的用户和当前用户的相关程度r。那么最终得到的权重为 E=u*w*d*r

对用户接收到的事件计算权重，排序和推荐可以提高用户体验，提升用户活跃度，增加用户使用微博的时间，进而增加广告等相关收入。
